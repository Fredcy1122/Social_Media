#### Replace unwanted character and change datatype for Engagement Rate (per Impression), Like and Reactions

```python
df['Engagement Rate (per Impression)'] = df['Engagement Rate (per Impression)'].str.replace('%', '')
df['Engagement Rate (per Impression)'] = df['Engagement Rate (per Impression)'].str.replace(',', '')
df['Engagement Rate (per Impression)'] = df['Engagement Rate (per Impression)'].astype(float)
df['Engagement Rate (per Impression)'].describe()
df['Likes'] = df['Likes'].str.replace(',', '')
df['Likes'] = df['Likes'].astype(float)
df['Reactions'] = df['Reactions'].str.replace(',', '')
df['Reactions'] = df['Reactions'].astype(float)
```

```python
# we encountered an error when we were  try to fill engagement rates with gorupby summary statistic,
# so it leads us to find the outliers on the engagement rates column
# first we identified rows with engagement rates getter than 100%
b = df['Engagement Rate (per Impression)'] > 100
# we filter out rows where Likes and Reactions are greater than Impressions
c = df['Impressions'] < df['Likes']
d = df['Impressions'] < df['Reactions']
df = df[~c & ~d]
# As a general rule, impressions tend to be higher than engagement. 
# This is because one person can see the same post more than once, 
# and not every person who sees your content is necessarily going to like it, 
# share it or comment on it. Goals that align more with increased impressions include elevated brand visibility 
# or increased brand awareness2. However, engagement increases as your impressions increase because your potential pool 
# of respondents is growing (-Google)

# but there are some rows were engagement rates is greater than impression because of summed likes, comments, click throughs etc
# so remove rows were engagement rates getter than 200%
e = df['Engagement Rate (per Impression)'] > 200
df = df[~e]
df.shape
```

```python
# apply replace type 1
df['Organic Reach'] = df['Organic Reach'].str.replace(',', '')
df['Organic Reach'] = df['Organic Reach'].astype(float)
oe_dic =df.groupby('Network')['Organic Reach'].median().replace(np.nan,0).to_dict()
df['Organic Reach'] = df['Organic Reach'].fillna(df['Network'].map(oe_dic))

# apply replace type 2
df['Viral Reach'] = df['Viral Reach'].str.replace(',', '')
df['Viral Reach'] = df['Viral Reach'].astype(float)
df['Viral Reach'] =df['Viral Reach'].fillna(0)
# apply replace type 1
ere_dic =df.groupby('Content Type')['Engagement Rate (per Impression)'].median().to_dict()
df['Engagement Rate (per Impression)'] = df['Engagement Rate (per Impression)'].fillna(df['Content Type'].map(ere_dic))

# apply replace type 1
df['Engagements'] = df['Engagements'].str.replace(',', '')
df['Engagements'] = df['Engagements'].astype(float)
ee_dic =df.groupby('Content Type')['Engagements'].median().to_dict()
df['Engagements'] = df['Engagements'].fillna(df['Content Type'].map(ee_dic))

# apply replace type 1
re_dic =df.groupby('Content Type')['Reactions'].median().replace(np.nan,0).to_dict()
df['Reactions'] = df['Reactions'].fillna(df['Content Type'].map(re_dic))

# apply replace type 1
le_dic =df.groupby('Content Type')['Likes'].median().replace(np.nan,0).to_dict()
df['Likes'] = df['Likes'].fillna(df['Content Type'].map(le_dic))

# apply replace type 1
df['Comments'] = df['Comments'].str.replace(',', '')
df['Comments'] = df['Comments'].astype(float)
ce_dic =df.groupby('Content Type')['Comments'].median().replace(np.nan,0).to_dict()
df['Comments'] = df['Comments'].fillna(df['Content Type'].map(ce_dic))
# apply replace type 1
df['Click-Through Rate'] = df['Click-Through Rate'].str.replace('%', '')
df['Click-Through Rate'] = df['Click-Through Rate'].astype(float)
ce_dic =df.groupby('Network')['Click-Through Rate'].median().replace(np.nan,0).to_dict()
df['Click-Through Rate'] = df['Click-Through Rate'].fillna(df['Network'].map(ce_dic))

# apply replace type 1
df['Post Link Clicks'] = df['Post Link Clicks'].str.replace(',', '')
df['Post Link Clicks'] = df['Post Link Clicks'].astype(float)
pe_dic =df.groupby('Network')['Post Link Clicks'].median().replace(np.nan,0).to_dict()
df['Post Link Clicks'] = df['Post Link Clicks'].fillna(df['Network'].map(pe_dic))

# apply replace type 1
df['Follows from Post'] = df['Follows from Post'].astype(float)
fe_dic =df.groupby('Network')['Follows from Post'].median().replace(np.nan,0).to_dict()
df['Follows from Post'] = df['Follows from Post'].fillna(df['Network'].map(fe_dic))
```

#### Exploratory and time intelligence analysis will be done using Power bi

Over the years, Stanbic PLC social media team has made efforts to increase the company's brand aware online with over 3,000 total post made on different social media every year.


The social media campaign was a success with an avagrage of 2440 impressions per year and an average of 62 engagements per year from 2013 to 2020 where it met its peak.
At 2021, there was a huge fall.

Both impression and engagements started declining

The Declining rate started in 2021 with 45%(impressions) and 60%(engagements) decrease compared to 2020.

2022 was strugling but 2023 is still experiencing this fall.

#### This is a problem because the campaign was set to increase the brand image and online presence for the business to be attract more customers, be competitive and profitable.

We are going to explore the cause of this problem and recommend a solution

We need to chose a metric to use for our analysis.

There is a moderately strong positive relationship between Impression and Engagement, therefore we will be using both the total and average(median) engagement for our analysis because the higher the engagements on our post, the higher impressions will go.

```python
df[['Impressions', 'Engagements']].corr()
```

Before we find the cause, we need to explore some categories to give us a proper understanding of our dataset.

- Top Social Media

Having exploring our data, we will now analyse the the 60% decrease of the sum of engagement between 2020 and 2021

Between 2020 and 2021, 

- Engagements for facebook and twitter decreased majorly
- These platforms are managed by Damilare and Sophia
- The content type of post that reduced these platforms engagement is Photo

Now we will use Multiple regression to predict our engagement for the next 3 year

```python
df.columns
```

```python
df['Sent by'].value_counts()
```

```python
df['Year'] = df['Date'].dt.year
```

```python
# first subset the data for Damilare, Sophia and Blessing
names = ['Blessing Ubah', 'Sophia Amuka', "Damilare Oyekanmi"]
content = ['Photo', 'Video', 'Text']
a = df['Sent by'].isin(names)
b = df['Content Type'].isin(content)
df5 = df[a & b]
df5 = df5[['Year', 'Network', 'Content Type', 'Sent by', 'Engagements']]
df5.head()
```

```python
df5['Network'] = df5['Network'].replace('Facebook', 1).replace('Instagram', 2).replace('Twitter', 3).replace('LinkedIn', 4)
df5['Content Type'] = df5['Content Type'].replace('Photo', 1).replace('Video', 2).replace('Text', 3)
df5['Sent by'] = df5['Sent by'].replace('Damilare Oyekanmi', 1).replace('Sophia Amuka', 2).replace('Blessing Ubah', 3)
df5.info()
```

```python
df5['Year2'] = df5['Year'] ** 2
```

```python
df5 = df5.rename(columns = {'Content Type':'Content_Type', 'Sent by':'Sent_by'})
```

```python
df5.head()
```

```python
import statsmodels.formula.api as smf
```

```python
model = smf.ols('Engagements ~ Year + Year2 + C(Network) + C(Content_Type) + C(Sent_by)' , data = df5).fit()
model.params
```

```python
dfp = pd.DataFrame()
dfp['Year'] = np.random.choice([2024, 2025, 2026, 2027], 10000)
dfp['Year2'] = dfp['Year'] ** 2
dfp['Network'] = 1
dfp['Content_Type'] = 1
dfp['Sent_by'] = 1
dfp.head()
```

```python
pred = model.predict(dfp)
dfp['pred'] = pred
```

```python
fb_ph = dfp.groupby('Year')['pred'].median()
fb_ph.plot(title = 'Engagement Prediction for the next 4 years', ylabel = 'average engagement')
```

the results from our prediction shows that average engagement is likely to decline to 0 at 2027
