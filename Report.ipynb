### Recommendation

From the analysis made, we recommend the following:

- Photo is our most employed content type and it has majority of the declining rate, we should distribute our post to other content type equally and we should revisit the graphics content of our photos
- Facebook is majorly manage by Damilare, other social media managers should involve more in managing facebook and twitter
- Majority of post is made on Facebook, other platforms like instagram and twitter should be optimised equally
- Audience are more engaging to weekend post especially sunday, we should utilize the weekends for posting on our accounts
- Audience are more active between 9pm to 1am, we should schedule our posting at those hours of the day
